```python
from mlxtend.frequent_patterns import apriori , association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
#sample dataset
dataset = [
    ['milk', 'bread', 'butter'],
    ['bread', 'butter', 'jam'],
    ['milk', 'bread', 'jam'],
    ['milk', 'bread', 'butter', 'jam'],
    ['butter', 'jam']
]
```

```python
dataset
```

```python
#step1 : preproccesing the data
te = TransactionEncoder()
te_array = te.fit_transform(dataset)#transform the dataset into binary matrix
df = pd.DataFrame(te_array, columns=te.columns_)#convert to a pandas df
```

```python
#step 2: genrate frequent itemsets using apirori algo
#min support is the threshold for t=itemset frequency 
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
```

```python
#step 3 genrate associatan rules
#calculate num_itemsets 
num_itemsets = len(frequent_itemsets.itemsets[0])

#now include num_itemset in the associationa rulws call
rules = association_rules(frequent_itemsets, metric="confidence" , min_threshold=0.7, num_itemsets = num_itemsets)
```

```python
#output results
print("Frequent itemsets: ")
print(frequent_itemsets)
print("\n Association Rules: ")
print(rules)
```

```python
import pandas as pd
import numpy as np
from apyori import apriori
     
```

```python
df=pd.read_csv('MBO.csv', header=None)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
#data cleaning step
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
#data pre processing step
# for using aprori , need to convert data in list format..
# transaction = [['apple','almonds'],['apple'],['banana','apple']]....

transaction = []
for i in range (0, len(df)):
    transaction.append([str(df.values[i,j]) for j in range(0, 20) if str(df.values[i,j]) !='0'])
```

```python
transaction[0]
```

```python
#verifying
transaction[1]
```

```python
# Call apriori function which requires minimum support, confidance and lift, min length is combination of item default is 2".

rules = apriori(transaction, min_support = 0.003, min_confidance= 0.2, min_lift = 3, min_length = 2)


## min_support = 0.003 -> means selecting items with min support of 0.3%
## min_confidance = 0.2 -> means min confidance of 20%
## min_lift = 3
## min_length = 2 -> means no. of items in the transaction should be 2
```

```python
#genrateas set of rules
rules
```

```python
#rules =need to converetd in list
Results = list(rules)
Results
```

```python
df_results = pd.DataFrame(Results)
```

```python
df_results.head()
```

```python
#keep support in a seprate fdata fraME  SO WE CAN USE LATER
support = df_results.support
```

```python
'''
convert orderstaticsit in a proper format
oredaer staticsti has lhs=> rhs as well as rhs =>lhs
we can choose any one fro convinvave

'''


#all 4 will contain lhs, rhs , confi amd lify
first_values = []
second_values = []
third_values = []
fourth_values = []
```

```python
#loop number of rows time and append 1 by 1 value in a separate list..
#first and second element was frozenset which need to be converted in list..



for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_values.append(single_list[3])
    
```

```python
#conver all four list in df for further operation
lhs = pd.DataFrame(first_values)
rhs = pd.DataFrame(second_values)

confidence = pd.DataFrame(third_values, columns=['Confidence'])
lift = pd.DataFrame(fourth_values , columns=['lift'])
```

```python
# concat all list together in a single dataframe
df_final = pd.concat([lhs,rhs,support, confidence,lift] ,axis =1)
df_final
```

```python
'''
 we have some of place only 1 item in lhs and some place 3 or more so we need to a proper
 represenation for User to understand.
 replacing none with ' ' and combining three column's in 1
 example : coffee,none,none is converted to coffee, ,
'''

df_final.fillna(value = ' ', inplace = True)
df_final.head()
```

```python
df_final.columns = ['lhs',1, 'rhs',2,3, 'support', 'confidence', 'lift']
df_final.head()
```
